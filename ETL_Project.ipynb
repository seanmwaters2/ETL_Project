```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

# Extracts CSV into DataFrames

```python
sales_file = "Resources/Sale_Prices_State.csv"
sales_df = pd.read_csv(sales_file)
sales_df.head()
```

```python
rental_file = "Resources/State_MedianRentalPrice_AllHomes.csv"
rental_df = pd.read_csv(rental_file)
rental_df.head()
```

# Cleaning the Data

```python
len(sales_df.columns)
sale_2019= sales_df.iloc[:, np.r_[2,134:146]]
sale_2019.head()
```

```python
len(rental_df.columns)
```

# Transform Rental DataFrame

```python
rental_avg = rental_df.iloc[:, np.r_[1,111:123]].mean(axis=1)
rental_avg
rental_df['Average'] = round(rental_avg,2)
rental_df.head()
rental_2019 = rental_df.iloc[:, np.r_[1,111:124]]
rental_2019.head()
```

# Transform Sales DataFrame

```python
sales_avg = sales_df.iloc[:, np.r_[2,134:146]].mean(axis=1)
#sales_df['Average'] = sales_avg
sales_df['Average'] = round(sales_avg,2)
#sales_df.head()
sales_2019 = sales_df.iloc[:, np.r_[2,134:146,149]]
sales_2019.head()
```

# Merging DataFrames

```python
combined_df = pd.merge(sales_2019, rental_2019, on= "RegionName")
combined_df.head()
```

```python
list(combined_df.columns)
```

```python
combined_df = combined_df[['RegionName', '2019-01_x','2019-01_y','2019-02_x','2019-02_y','2019-03_x','2019-03_y','2019-04_x','2019-04_y',\
                        '2019-05_x','2019-05_y','2019-06_x','2019-06_y','2019-07_x','2019-07_y','2019-08_x', '2019-08_y','2019-09_x','2019-09_y', \
                         '2019-10_x','2019-10_y','2019-11_x','2019-11_y','2019-12_x','2019-12_y','Average_x','Average_y']]
combined_df.head()
```

```python
list(combined_df.columns)
```

```python
combined_df.columns = ['RegionName', '2019-Jan-Sales', '2019-Jan-Rental', '2019-Feb-Sales', '2019-Feb-Rental', '2019-Mar-Sales', '2019-Mar-Rental',\
                      '2019-Apr-Sales','2019-Apr-Rental', '2019-May-Sales', '2019-May-Rental', '2019-Jun-Sales','2019-Jun-Rental',\
                       '2019-July-Sales', '2019-July-Rental', '2019-Aug-Sales', '2019-Aug-Rental', '2019-Sept-Sales', '2019-Sept-Rental', \
                      '2019-Oct-Sales', '2019-Oct-Rental', '2019-Nov-Sales', '2019-Nov-Rental', '2019-Dec-Sales', '2019-Dec-Rental', 'Sales-Avg','Rental-Avg']
combined_df.head()
```
